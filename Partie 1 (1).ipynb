###  2.- Question 1

### Quels sont les 10 articles ayant le plus d’avis ?
### Afficher graphiquement leur nombre d’abonnés par ordre décroissant.

```python
# Pour répondre à cette question on va procéder au tri du Dataframe data_artists par ordre décroissant 
#en fonction des variables popularity et followers
data_artists.sort_values(["popularity","followers"], ascending=[False,False],inplace=True)
data_artists["followers_M"] = round(data_artists["followers"] / 1000000, 2)

```

```python
most_pop=data_artists.head(10)
most_pop
```

### Représentation graphique

```python
plt.figure(figsize=(10, 6))
plt.barh(most_pop["name"], most_pop["followers_M"],color=plt.cm.gist_rainbow(np.linspace(0, 1, len(most_pop["name"]))))
plt.gca().invert_yaxis()
plt.xlabel("Nom des artistes")
plt.ylabel("Nombre d'abonnés en millions")
plt.title("artistes par nombre d'abonnés ")
plt.suptitle("Source:data_artists")
plt.xticks(rotation=90)
plt.show()

```
